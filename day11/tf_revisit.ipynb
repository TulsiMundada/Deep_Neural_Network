```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import tensorflow as tf

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.preprocessing import StandardScaler , LabelEncoder

import warnings
warnings.filterwarnings("ignore")
```

```python
RANDOM_STATE = 24

np.random.seed(RANDOM_STATE)
tf.random.set_seed(RANDOM_STATE)

EPOCHS = 200
TEST_SIZE = 0.2
```

```python
data_df = pd.read_csv("D:\DNN\dnn_lab\day11\wine.csv", header=None)
data_df
```

```python
labels = data_df[data_df.columns[-1]]
labels
```

```python
features_df = data_df.drop(data_df.columns[-1],axis=1)
features_df
```

```python
X_train , X_test ,y_train , y_test = train_test_split(features_df , labels ,stratify = labels ,
                                                      test_size = TEST_SIZE , random_state = RANDOM_STATE)
X_train.shape , X_test.shape ,y_train.shape , y_test.shape
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
```

```python
model = tf.keras.Sequential()

model.add(tf.keras.layers.Dense(8 ,
    activation='relu',
    kernel_regularizer=None,))
model.add(tf.keras.layers.Dense(3))
```

```python

loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)

model.compile(loss=loss_fn , optimizer='adam' , metrics=['accuracy'])
```

```python
hist = model.fit(X_train  , y_train , epochs=EPOCHS , validation_data=[X_test , y_test])
```

```python
loss_df = pd.DataFrame(hist.history)
loss_df.head()
```

```python
loss_df.plot(y=['loss','val_loss'])
```

```python
loss_df.plot(y=['accuracy','val_accuracy'])
```

```python
y_pred = model.predict(X_test)
y_pred.shape
```

```python
y_pred = y_pred.argmax(axis=1)
```

```python
accuracy_score(y_test , y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test , y_pred)
```
